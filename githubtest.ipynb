```python
import base64

def fetch_directory_tree_recursive(dir_url, headers):
    """Recursively fetch the directory tree and extract file contents."""
    tree = {}
    try:
        dir_response = requests.get(dir_url, headers=headers)
        dir_response.raise_for_status()
        dir_contents = dir_response.json()
        
        for item in dir_contents:
            if item['type'] == 'dir':
                # Recursively fetch the subdirectory
                tree[item['name']] = fetch_directory_tree_recursive(item['url'], headers)
            elif item['type'] == 'file':
                # Fetch the file content
                try:
                    file_response = requests.get(item['url'], headers=headers)
                    file_response.raise_for_status()
                    file_content = file_response.json().get('content')
                    if file_content:
                        # Decode the base64 content
                        tree[item['name']] = base64.b64decode(file_content).decode('utf-8')
                except Exception as e:
                    print(f"Failed to fetch file content for {item['name']}: {e}")
    except Exception as e:
        print(f"Failed to retrieve directory at {dir_url}: {e}")
    return tree

# Start building the tree from the root directory
root_dir_url = organization_repositories_url  # Use the root directory URL
directory_tree = fetch_directory_tree_recursive(root_dir_url, headers)

print(directory_tree)
```

```python
import base64
import time
from requests.exceptions import ConnectionError, HTTPError

def fetch_with_retries(url, headers, retries=3, delay=2):
    """Fetch a URL with retries and fixed delay."""
    for attempt in range(retries):
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  # Raise HTTPError for bad responses
            return response
        except (ConnectionError, HTTPError) as e:
            if attempt < retries - 1:
                print(f"Retrying in {delay} seconds... (Attempt {attempt + 1}/{retries})")
                time.sleep(delay)
            else:
                print(f"Failed to fetch {url} after {retries} attempts.")
                raise e

def fetch_directory_tree(dir_url, headers):
    tree = {}
    try:
        dir_response = fetch_with_retries(dir_url, headers)
        dir_contents = dir_response.json()
        for item in dir_contents:
            if item['type'] == 'dir':
                # Recursively fetch the subdirectory
                tree[item['name']] = fetch_directory_tree(item['url'], headers)
            elif item['type'] == 'file':
                # Fetch the file content
                try:
                    file_response = fetch_with_retries(item['url'], headers)
                    file_content = file_response.json().get('content')
                    if file_content:
                        # Decode the base64 content
                        tree[item['name']] = base64.b64decode(file_content).decode('utf-8')
                except Exception as e:
                    print(f"Failed to fetch file content for {item['name']}: {e}")
    except Exception as e:
        print(f"Failed to retrieve directory at {dir_url}: {e}")
    return tree

# Start building the tree from the root directory
root_dir_url = organization_repositories_url  # Use the root directory URL
directory_tree = fetch_directory_tree(root_dir_url, headers)

print(directory_tree)
```

```python
import tiktoken

def estimate_tokens(text, model="gpt-3.5-turbo"):
    """Estimate the number of tokens in a given string for a specific model."""
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)
    return len(tokens)

# Example usage
# Read the content of the text file
with open("/Users/akhilred/Downloads/cbs.txt", "r") as file:
    text = file.read()
num_tokens = estimate_tokens(text)
print(f"Estimated number of tokens: {num_tokens}")
```

```python
from tree_sitter import Language, Parser
import os

# Build the language library if not already built
Language.build_library(
    'build/my-languages.so',
    [
        'vendor/tree-sitter-python'  # Path to the language grammar
    ]
)

# Load the language (assuming Python for this example)
PY_LANGUAGE = Language('build/my-languages.so', 'python')
parser = Parser()
parser.set_language(PY_LANGUAGE)


def parse_repo_tree(repo_path, file_extension=".py"):
    tree_dict = {}
    for root, dirs, files in os.walk(repo_path):
        for file in files:
            if file.endswith(file_extension):
                file_path = os.path.join(root, file)
                with open(file_path, "rb") as f:
                    code = f.read()
                tree = parser.parse(code)
                tree_dict[file_path] = tree
    return tree_dict

# Example usage:
repo_path = "/Users/akhilred/Desktop/gittest/cloud-backup-service"
repo_trees = parse_repo_tree(repo_path)
print(repo_trees)
```
